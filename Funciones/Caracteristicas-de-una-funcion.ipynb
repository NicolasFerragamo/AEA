## <span style="color:red">  Dominio </span>

<div class="alert alert-block alert-info">
<b> Definición: </b>
    
El <b> dominio </b> de una función $f: A \subseteq \mathbb{R} \rightarrow  B \subseteq \mathbb{R}$, es el conjunto de <b>x</b> para los cuales es válida la regla de correspondencia <b>f</b>
</div> 

El dominio son todos los valores de **A** que que estan incluidos ($\subseteq$) en los números **Reales**. No todos los valores son validos, ya que dependen de la función. Ahora con los siguientes ejemplos esto va a quedar mas claro. 

<div class="alert alert-block alert-success">
<b> Ejemplo n° 1: </b>
    
Dada la función $f(x)=x^2$. ¿hay algun valor de <b>x</b> que no este permitido?

$$$$
    
Para esta función llamada <b> función cuadrática</b>, todos los valores son posibles.
    
$$$$
    
Matemáticamente se escribe como:
    
$$D_f = \mathbb{R}$$    
    
Esto quiere decir que el dominio de <b> f</b> son todos los números reales.    
</div> 




```python
# nbi:hide_in
%matplotlib inline
plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})
x  = np.linspace(-20,20,400)

plt.figure("y=f(x)=x²")
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(x,x**2,color='red', label='f(x)=x²')
plt.grid()
plt.title('y=f(x)=x² ')
plt.legend(loc = 'upper center')
plt.show()
           
```

<div class="alert alert-block alert-success">
<b> Ejemplo n° 2: </b>
    
Dada la función $f(x)=\frac{1}{x-4}$. ¿hay algun valor de <b>x</b> que no este permitido?

$$$$
    
En este caso tenemos que tener cuidado con los valores que toma la <b>x</b> ya que debemos recordar que no se puede dividir por "0". Por lo cual procedemos a evaluar que valores de <b>x</b> hace que el denominador se anule. 
    
    
$$x - 4 = 0$$
    
$$x = 4$$ 
    
Por lo cual el dominio son todos los números reales menos el 4. Esto formalmente se escribe:    
    

$$D_f = \mathbb{R}- 4$$    
    
Otra forma de escribirlo sería    
    
$$D_f = \{ x/ x \in \mathbb{R} \land x\ne 4 \}$$      
</div> 


### <span style="color:brown">Podemos notar en el gráfico que la función no existe en 4.</span>

```python
# nbi:hide_in
np.seterr(divide='ignore', invalid='ignore') # para eliminar el carter de división por 0

x = np.linspace(-5,15,1000)
y = 1/(x - 4)

plt.figure("y=f(x)=1/(x - 4)")
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(x,y,color='red', label='f(x)=1/(x - 4)')
plt.grid()
plt.title('y=f(x)=1/(x - 4) ')
plt.legend(loc = 'upper right')
plt.show()
```

<div class="alert alert-block alert-success">
<b> Ejemplo n° 3: </b>
    
Dada la función $f(x)=\sqrt{x+4}$. ¿hay algun valor de <b>x</b> que no este permitido?

$$$$
    
Recordemos que no se puede obtener la raiz cuadrada de un número menor a "0". Por lo cual procedemos a evaluar que valores de <b>x</b> hace que el radicando sea mayor o igual a 0. 
    
    
$$x + 4 \ge 0$$
    
$$x \ge -4$$ 
    
En este caso el dominio son todos los números reales mayores o igual a -4. Esto formalmente se escribe:    
    
$$$$ 
    
$$D_f = \{ x/ x \in \mathbb{R} \land x\ge -4 \}$$    
    
$$$$
    
Otra forma de escribirlo sería:
    
$$$$
    
$$D_f= [-4; +\infty)$$     
    
</div> 

```python
# nbi:hide_in
x = np.linspace(-4,100,1000)
y = np.sqrt(x + 4)


plt.figure("y=f(x)=sqrt(x + 4)")
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(x,y,color='red', label='f(x)=sqrt(x + 4) ')
plt.grid()
plt.title('y=f(x)=sqrt(x + 4) ')
plt.legend(loc = 'upper left')
plt.show()
```

## <span style="color:red">  Codominio </span>

<div class="alert alert-block alert-info">
<b> Definición: </b>
    
El <b> codominio </b> de una función $f: A \subseteq \mathbb{R} \rightarrow  Y \subseteq B$, es el conjunto <b>B</b> que participa en la función. Se denomina $Cod_f$.
    

Estrictamente es el subespacio vectorial del conjunto de llegada. (pero esta definición escapa del alcance de este apunte)
</div> 

En otras palabras es el conjunto de todos lo valores que se encuentran en el conjunto de llegada, mas aya que los valores sean o no tomados por la función.


## <span style="color:red">  Imagen </span>

<div class="alert alert-block alert-info">
<b> Definición: </b>
    
El <b> Imagen </b> de una función $f: A \subseteq \mathbb{R} \rightarrow  Y \subseteq B$, es el conjunto <b>Y</b> contenido en <b>B</b> (es un subconjunto de Y) formado por todos los valores que puede tomar la función. Se denota $Im_f$.
    
$$$$
    
$$Im_f= \{y \in Y  / \exists x \in A \land f(x)=y\}$$    


</div> 

Esto es muy formal, pero significa lo siguiente: los valores de **y** (los valores que devuelve la función para cada **x**) que pertenecen al **codominio**, tal que existe un **x** que pertenece al **dominio** y $f(x)=y$ (le asignamos a **y** un valor de acuerdo a la regla).

Esto es solo informativo, lo volverán a ver detalladamente en matemática.

Veamos algunos ejemplos para que se entienda el concepto.

<div class="alert alert-block alert-success">
<b> Ejemplo n°1: </b>
    
Para el ejemplo n°1 visto antes $f(x)=x^2$, podemos ver en el gráfico que los valores de <b> y</b> que toma la función van de $0$ a $+\infty$. El codominio son todos los valores del conjunto de llegada, en este caso son todos los números reales. Escribamos esto:
    
$$$$
    
$$Cod_f = \mathbb{R}$$  
    
$$$$    

$$Im_f = \{ y \in \mathbb{R} / y=x^2 \land x \in \mathbb{R} \}$$
    
$$$$
    
Podemos escribír esto mas fácilmente como: 
    
$$$$
    
$$ Im_f = [0; +\infty) $$    
</div> 



<div class="alert alert-block alert-success">
<b> Ejemplo n°2: </b>
    
Para el ejemplo n°2 visto antes, $f(x)=\frac{1}{x-4}$, podemos ver en el gráfico que los valores de <b> y</b> que toma la función todos los número reales excepto el $4$. El codominio son todos los valores del conjunto de llegada, en este caso son todos los números reales. Escribamos esto:
    
$$$$
    
$$Cod_f = \mathbb{R}$$  
    
$$$$    

$$Im_f = \{ y \in \mathbb{R} / y=\frac{1}{x-4} \land x \neq 4 \}$$
    
$$$$
    
Podemos escribír esto mas fácilmente como: 
    
$$$$
    
$$ Im_f = \mathbb{R}- \{4\} $$    
</div> 

Por último veamos el tercer ejemplo.

<div class="alert alert-block alert-success">
<b> Ejemplo n°3: </b>
    
Para el ejemplo n°3 visto antes, $f(x)=\sqrt{x+4}$, podemos ver en el gráfico que los valores de <b> y</b> que toma la función van de $0$ a $+\infty$. El codominio son todos los valores del conjunto de llegada, en este caso son todos los números reales. Escribamos esto:
    
$$$$
    
$$Cod_f = \mathbb{R}$$  
    
$$$$    

$$Im_f = \{ y \in \mathbb{R} / y=\sqrt{x+4} \land x \ge 4 \}$$
    
$$$$
    
Podemos escribír esto mas fácilmente como: 
    
$$$$
    
$$ Im_f = [0; +\infty) $$    
</div>

## <span style="color:red">  Gráfica de una función </span>

<div class="alert alert-block alert-info">
<b> Definición: </b>
    
Sea una función $f: A \subseteq \mathbb{R} \rightarrow  B \subseteq \mathbb{R}$, la <b> gráfica</b> de $f$ es el conjunto de pares ordenados $\{ (x,f(x))/ x \in A \}$
</div> 

Dejando el formalismo un poco de lado, esto quiere decir que la <span style="color:red"> <b>gráfica</b> </span> de $f$ es el conjunto de pares ordenados $(x,y)$ (los puntos formados por los valores de $x$ e $y$, ordenados en este orden) tal que $y=f(x)$ ($y$ es la imagen de la función para cada valor de $x$). En otras palabras es la gráfica de la ecuación $y=f(x)$. Podemos pensar al valor de $f(x)$ como la altura de la gráfica en el punto $x$.

La gráfica de una función nos muestra el valor de la imagen de la función para cada valor del dominio, por lo cual nos da mucha información, como los valores posibles de $x$ e $y$. Nos muestra como es su forma, si crece, si decrece, su valor máximo y mínimo si los tiene, su intersección con los ejes coordenados, etc. En otras palabras, nos da un cuadro sobre el comportamiento de la función.

Conocer esto nos va a ser de suma importancia al momento de graficar señales y al momento de visualizarlas.

En esta sección vimos muchas gráficas al momento de obtener el dominio, codominio e imagen de una función.

### <span style="color:red">  Como graficar una función </span>

Veamos como obtener la gráfico de una función. Para ello podemos realizar una tabla con algunos valores, luego dibujar los puntos en un eje carteciano y por último unir los puntos. Es parecido al juego donde teníamos que unir los puntos en un orden determinado.

La mejor forma de aprender esto es resolviendo algunos ejemplos.

<div class="alert alert-block alert-success">
<b> Ejemplo n°4 </b>
    
Grafique la  función $y = 2x+1$

$$$$

Lo primero que hacemos es armar una tabla de valores, en la cual vamos a elegir los valores de $x$ y vamos a remplazarlo en la ecuación para obtener el valor de $y$.   
    
Por ejemplo para $x=0$ nos queda $y=2*0+1=1$, para $x=1$ nos queda $y=2*1+1=3$ y así para los valores que elijamos.
</div>

|:-:  | :------: |

<div class="alert alert-block alert-success"> 
Ahora procedemos a marcar los puntos en un sistema de ejes coordenados.   
</div> 

```python
# nbi:hide_in

np.seterr(divide='ignore', invalid='ignore') # para eliminar el carter de división por 0
plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})

p_x = np.array([-10, -5, -3, -1, 0, 1, 4, 10])
p_y = np.array([-19, -9, -5, -1, 1, 3, 9, 21])

texto = ['(-10;-19)', '(-5;-9)', '(-3;-5)', '(-1;-1)', '(0;1)', '(1;3)', '(4;-9)', '(10;21)']


plt.figure("y=f(x)=2x+1",figsize=(7, 7), dpi= fig_dpi, facecolor='w', edgecolor='k')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(p_x,p_y,color='red', label='f(x)=y = 2x+1', marker='o', linestyle='None')
plt.grid()
plt.title('y=f(x)=y = 2x+1 ')
plt.legend(loc = 'best')
plt.axhline(0, color="grey")
plt.axvline(0, color="grey")
    
for ii in range(len(p_x)):
    plt.text(p_x[ii],p_y[ii],texto[ii],fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
             horizontalalignment='left', verticalalignment='bottom')
        
plt.show()
```

<div class="alert alert-block alert-success"> 
Por último procedemos a unir los puntos   
</div>

```python
# nbi:hide_in
    
plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})

p_x = np.array([-10, -5, -3, -1, 0, 1, 4, 10])
p_y = np.array([-19, -9, -5, -1, 1, 3, 9, 21])

texto = ['(-10;-19)', '(-5;-9)', '(-3;-5)', '(-1;-1)', '(0;1)', '(1;3)', '(4;-9)', '(10;21)']
x = np.linspace(-10,10,1000)
y = 2*x+1


plt.figure("y=f(x)=2x+1",figsize=(7, 7), dpi= fig_dpi, facecolor='w', edgecolor='k')
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.plot(p_x,p_y,color='red', label='f(x)=y = 2x+1', marker='o', linestyle='None')
plt.plot(x,y,color='blue')
plt.xlim(min(x)-0.2, max(x)+0.2)
plt.ylim(min(y)-0.2, max(y)+0.2)
plt.grid()
plt.axhline(0, color="grey")
plt.axvline(0, color="grey")
plt.title('y=f(x)=y = 2x+1 ')
plt.legend(loc = 'best')
    
for ii in range(len(p_x)):
    plt.text(p_x[ii],p_y[ii],texto[ii],fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')

plt.show()


```

## <span style="color:red">  Intersección con los ejes coordenados </span>

Otro de los temas importantes a tener en cuenta es la intersección de la gráfica con los ejes coordenados. Esto simplemente es el punto donde la gráfica corta al eje $x$ y donde al eje $y$. Esto nos va a permitir realizar las gráficas mas fácilmente.

<div class="alert alert-block alert-info">
<b> Definición de ceros: </b>
    
Se denomina <b>ceros</b> de la función, a los puntos donde la función <b>intersectar</b> al <b>eje de abscisas</b> (el eje$ x$).
    
$$$$ 
    
$a$ es cero de $f \Leftrightarrow f(a)=0$    
</div> 

Esto quiere decir que $a$ es **cero** de la función $f$ **si y solo si** $f$ evaluada en $a$ (en el punto $x=a$) es igual a $0$. 

También se lo llama **abscisas al origen** ya que son los puntos del eje de abscisas para los cuales el valor del eje de ordenadas (eje $y$ es **cero**.

<div class="alert alert-block alert-info">
<b> Definición de ordenada al origen: </b>
    
Se denomina <b>ordenada al origen</b> a el punto donde la función <b>intersectar</b> al <b>eje de ordenadas</b> (el eje$ y$).
    
$$$$ 
    
$b$ es la ordenada al origen de $f \Leftrightarrow f(0)=b$    
</div> 

Esto quiere decir que $b$ es la **ordenada al origen** de la función $f$ **si y solo si** $f$ evaluada en $0$ (en el punto $x=0$) es igual a $b$. No es otra cosa que el punto donde la función intersecta al $eje$ $y$

Veamos un ejemplo para entenderlo mejor.


<div class="alert alert-block alert-success">
<b> Ejemplo n° 6: </b>

Obtener la intersección de la función con los ejes coordenados.

$$ y = f(x) = x^2 - 3$$

Lo que pide el ejercicio es que obtengamos los <b>ceros</b> y la <b> ordenada al origen</b>

Vamos a graficar la función para verlos visualmente y luego los obtenemos analíticamente.

## <span style="color:red">  Simetría </span>

En algunos casos puede simplificarse la realización de la gráfica de una función teniendo en cuenta la **simetría**. Por ejemplo en la gráfica de arriba podemos ver que la función es simétrica con respecto al $eje$ $y$. Es como poner un espejo en el $eje$ $y$ y ver el reflejo de la función. En otros casos es simétrica con respecto al origen de coordenadas (el punto (0;0)) como el caso de la recta del ejemplo n° 4.

<div class="alert alert-block alert-info">
<b> Definición de paridad "par": </b>
    
Una función $f$ es <b>par</b> $\Leftrightarrow f(x) = f(-x), \forall x \in D_f$ 

</div> 

Es decir una función es **par** si la imagen de $f(x)=f(-x)$. Es **simétrica** con respecto al **eje** de **ordenadas**.

Podemos ver en la gráfica del ejemplo n° 6, que el valor de $f$ para $x=2$ es el mismo que para $x=-2$, lo mismo sucede para $x=-1$ y $x=1$, lo mismo sucede para cualquier valor de $x$. Por lo cual si graficamos solo para $x \ge 0$, podemos obtener el resto gel gráfico simplemente reflejando esta porción del otro lado del $eje$ $y$.

<div class="alert alert-block alert-info">
<b> Definición de paridad "impar": </b>
    
Una función $f$ es <b>impar</b> $\Leftrightarrow f(x) = -f(-x), \forall x \in D_f$ 

</div> 

Es decir una función es **impar** si la imagen de $f(x)=-f(-x)$. Es **simétrica** con respecto al **orige** de **coordenadas**, es decir **el punto (0;0)**.

Podemos ver en la gráfica del ejemplo n° 4, que el valor de $f$ para $x=2$ es el mismo que para $x=-2$ pero con signo invertido, lo mismo sucede para $x=-1$ y $x=1$, lo mismo sucede para cualquier valor de $x$. Por lo cual si graficamos solo para $x \ge 0$, podemos obtener el resto gel gráfico simplemente girando $180°$ respecto al origen (esto equivale a reflejar primero en el $eje$ $x$ y luego en el $eje$ $y$.


La siguiente imagen ilustra esto. Note que las distancias **a** son iguales y lo mismo ocurre con la distancias **b**. (solo son iguales en el mismo gráfico)

```python
# nbi:hide_in
x  = np.linspace(-5,5,1000)
y  = x**2
y1 = x**3

#grafico de las ventanas
fig = plt.figure(figsize=(10, 5), dpi= fig_dpi , constrained_layout=True)
gs = fig.add_gridspec(1, 2)

#grafico de la ventana de python
f_ax1 = fig.add_subplot(gs[0, 0])
f_ax1.set_title("Función par ")
f_ax1.plot(x,y,label='y = f(x) = x²')
f_ax1.set_xlabel("x")
f_ax1.set_ylabel('y=f(x)',rotation='horizontal')
f_ax1.arrow(-3.162, 10, 6.324, 0,color='r',length_includes_head=True, shape='full')
f_ax1.arrow( 3.162, 0, 0, 10,color='g',length_includes_head=True, shape='full')
f_ax1.arrow(-3.162, 0, 0, 10,color='g',length_includes_head=True, shape='full')
f_ax1.text(-1.5,11,'a',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')
f_ax1.text(1.5,11,'a',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')
f_ax1.text(-3.8,5,'b',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')
f_ax1.text(3.5,5,'b',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')

f_ax1.axhline(0, color='grey')
f_ax1.axvline(0, color='grey')
f_ax1.grid()
f_ax1.legend(loc = 'best')

#grafico del modulo la ventana de python
f_ax2 = fig.add_subplot(gs[0, 1])
f_ax2.set_title("Función impar " )
f_ax2.plot(x,y1, label='y = f(x) = x³')
f_ax2.set_xlabel('x')
f_ax2.set_ylabel('y=f(x)',rotation='horizontal')
f_ax2.arrow(-4, -64, 8, 128,color='r',length_includes_head=True, shape='full')
f_ax2.arrow( 4, 0, 0, 64,color='g',length_includes_head=True, shape='full')
f_ax2.arrow(-4, 0, 0, -64,color='g',length_includes_head=True, shape='full')
f_ax2.text(-2,-50,'a',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom', rotation=34)
f_ax2.text(1.8,45,'a',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom', rotation=34)
f_ax2.text(-4.6,-30,'b',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')
f_ax2.text(4.3,30,'b',fontsize=12,bbox=dict(facecolor='red', alpha=0.2),
                horizontalalignment='left', verticalalignment='bottom')
f_ax2.axhline(0, color='grey')
f_ax2.axvline(0, color='grey')
f_ax2.grid()
f_ax2.legend(loc='upper center')
```
