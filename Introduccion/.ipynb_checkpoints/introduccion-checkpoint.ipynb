# <span style="color:blue"> Introducción <span>

Tal vez se pregunten ¿Por qué estudiar electrónica analogía en la era digital? La realidad es que la digitalización de la electrónica trajo grandes beneficios  y permite un avance mas acelerado de la tecnología, pero... la electrónica digital tiene sus limitaciones aunque a priori no lo parezca. Por ejemplo en las aplicaciones donde se necesita trabajar con grandes potencias de energía hay que utilizar la electrónica analógica. Otro caso es cuando se necesita una gran velocidad para el tratamiento de una señal, la electrónica digital empieza a fallar. Si bien es verdad que la electrónica analógica fue desplazada en ciertos sectores, en el día a día conviven ambas. No podemos dejar de recordar que nuestros sentidos son analógicos, y la señales en la naturaleza también lo son, por lo cual tenemos que convertir esas señales de analógicas a digitales y viceversa y para ello usamos electrónica analógica. Desde luego que esto es solo un ejemplo y su utilización no termina acá.    

Esta materia esta relacionada de forma directa con **Física, Matemática, Fundamentos de los modelos circuitales, Montaje de proyectos Electrónico, Aplicaciones de electrónica Digital y Tecnología electrónica** de este año y con las correspondiente materias técnicas del resto de la carrera.

## <span style="color:blue"> Conocimientos previos<span>

Los conocimientos previos no son muchos, solo conocimientos básicos de matemática y un poco de física y química.

<div class="alert alert-block alert-info">
 <b> ¿Por qué debo saber matemática? </b>
</div>
La rama técnica se basa en los conocimientos de las ciencias básicas que son la matemática, la física y la química. Estas últimas utilizan el lenguaje de la matemática para desarrollar modelos que nos permitan explicar y predecir ciertos fenómeno. Como derivados de estas áreas también usamos modelos matemáticos para entender y diseñar circuitos eléctricos y electrónicos.

<div class="alert alert-block alert-info">
 <b> ¿Qué debo saber de matemática? </b>
</div>

Para esta materia los conocimientos no son muchos, solo se necesita saber sumar, restar, multiplicar, dividir, potenciación, radicación y usar fracciones no esta de mas. Pero sobre todo es fundamental saber **despejar ecuaciones.** Tras tres años impartiendo esta materia, me di cuenta que esta es una de las principales dificultades que tienen los estudiantes, por lo que les sugiero practicar esta habilidad. 

<div class="alert alert-block alert-danger">
 <b> ¿Por qué debo saber física y química? </b>
</div>

La física y la química explican por que ocurren los fenómenos eléctricos y nos dan los modelos matemáticos que utilizamos para poder predecir resultados, los cuales vamos a utilizar para poder diseñar nuestros circuitos electrónicos.

<div class="alert alert-block alert-danger">
 <b> ¿Qué debo saber de física y química? </b>
</div>
