# <span style="color:blue">  Trigonometría </span>

## <span style="color:blue">  Introducción </span>

En esta sección estudiaremos algunas funciones trigonométricas, su origen y su interpretación, sus características. En particular vamos a trabajar con la función **senoidal** que va ser de suma importancia en el estudio de **señales** ya que aparece de forma natural en muchos ámbitos de la naturaleza. Sin ir mas lejos la variación de la diferencia de potencial que llega a nuestros hogares, varia de forma **senoidal**. Las transmisión de señales, de radio o televisión, siguen patrones **senoidales**. La variación de la posición de un péndulo o un oscilador armónico simple siguen el mismo patrón. Las notas musicales también. Como habrán notado la función senoidal esta en todos lados y es nuestro objetivo entenderlas para poder usarlas.


## <span style="color:red">  Círculo unitario </span>

Vamos a analizar algunas propiedades del circulo de **radio** $1$ con centro en el origen de coordenadas, que nos van a permitir definir a las funciones trigonométricas.

<div class="alert alert-block alert-info">
<b> Definición de círculo unitario:</b> 
Es aquel aquel que tiene un radio iguala a $1$ y su centro está e el origen de un plano $xy$. Su ecuación es:
    
$$x^2 + y^2 = 1$$    
</div> 

Nota: el círculo no es una función porque no cumple el requisito de **unicidad**.

```python
# nbi:hide_in
plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})
plt.style.use('seaborn-pastel')

angulo = np.linspace(0, 2*np.pi, 1000)
x = np.cos(angulo) 
y = np.sin(angulo) 

plt.figure('Circulo uitario',figsize=(6, 6), dpi= fig_dpi, facecolor='w', edgecolor='k')
plt.title("Circulo Unitario")
plt.plot(x, y, markersize=1, label='x² + y² = 1')
plt.axhline(0, color='grey')
plt.axvline(0, color='grey')
plt.xlabel("x")
plt.ylabel("y",rotation='horizontal')
plt.plot(1, 0, 'ro')
plt.plot(0, 1,'ro')
plt.plot(-1, 0, 'ro')
plt.plot(0, -1, 'ro')
plt.arrow(0, 0, 1/np.sqrt(2)-0.07, 1/np.sqrt(2)-0.07, head_width=0.05, head_length=0.1, fc='r', ec='r')
plt.text(1.1,0,'$(0,0)$',fontsize=12,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='center')
plt.text(0,1.1,'$(0,1)$',fontsize=12,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='center', verticalalignment='bottom')
plt.text(-1.1,0,'$(-1,0)$',fontsize=12,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='right', verticalalignment='center')
plt.text(0,-1.1,'$(0,-1)$',fontsize=12,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='center', verticalalignment='top')
plt.text(0.44, 0.4,'$R=1$',fontsize=12,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='center', verticalalignment='top', rotation=45)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.grid()
plt.show()

```

Preste atención a los puntos marcados en el gráfico, nos van a ser de utilidad mas adelante.

<div class="alert alert-block alert-success">
<b> Ejemplo n° 1:</b> 
Demuestre que el punto $P(\frac{\sqrt(3)}{3},\frac{\sqrt(6)}{3})$ esta en el círculo unitario.
$$$$
Para demostrar que este punto cumple con la ecuación con la ecuación del círculo unitario, $x^2 + y^2 = 1$, debemos simplemente remplazar ese punto en la ecuación y si el resultado es $1$ como dice la ecuación, el punte pertenece, de lo contrario no.
    
$$(\frac{\sqrt{3}}{3})^2,(\frac{\sqrt{6}}{3})^2 = \frac{3}{9} + \frac{6}{9} = 1$$    
</div>

<div class="alert alert-block alert-success">
<b> Ejemplo n° 2:</b> 
El punto $P(\frac{\sqrt(3)}{2},y)$ esta en el círculo unitario en el cuadrante IV. Encuentre su coordenada $y$.
$$$$
Como sabemos que esta en el círculo unitario, el punto cumple la ecuación $(\frac{\sqrt{6}}{3})^2 + y^2 = 1$, debemos simplemente despejar el valor de $y$.
    
$$(\frac{\sqrt{6}}{3})^2 + y^2 = 1$$    
    
$$\frac{3}{4} + y^2 = 1$$    
    
$$y^2 = 1 -\frac{3}{4}  $$      
    
$$y = \sqrt{\frac{1}{4}}$$  
    
$$y = \pm \frac{1}{2}$$    
    
$$$$
    
Note que hay 2 resultados posibles, esto es muy importante. Como nos da la información adicional que esta en el cuarto cuadrante, el resultado final es   
    
$$\boxed{y = -\frac{1}{2}}$$  
</div>

Estos dos ejemplos están para mostrar como moverse sobre el círculo. Profundizaremos esto de inmediato.

Suponga que <span style="color:red"> <b> $t$ </b> </span> es un número real. Recorreremos una distancia <span style="color:red"> <b> $t$ </b> </span> a lo largo del círculo unitario, empezando en el punto $P(1,0)$ y desplazándonos en sentido antihorario si <span style="color:red"> <b> $t > 0$</b> </span> y en sentido horario si <span style="color:red"> <b> $t < 0$</b> </span>.

```python
# nbi:hide_in
%matplotlib inline

plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})


num_segmentos = 200
rad = 1
cx = 0
cy = 0

angulo = np.linspace(0, 2*np.pi, num_segmentos+1)
x = rad * np.cos(angulo) + cx
y = rad * np.sin(angulo) + cy

angulo = np.linspace(0, (2/3)*np.pi, num_segmentos+1)
x1 = rad * np.cos(angulo) + cx
y1 = rad * np.sin(angulo) + cy


#grafico de las las funciones
fig = plt.figure(figsize=(10,5), dpi= fig_dpi , constrained_layout=True)
gs = fig.add_gridspec(1, 2)

#grafico de la función original
f_ax1 = fig.add_subplot(gs[0, 0])
f_ax1.set_title("$Sentido$ $antihorario$ $t > 0$ ")
f_ax1.plot(x, y, color="red", markersize=1)
f_ax1.plot(x1, y1, color="b", markersize=1)
f_ax1.plot(1, 0, "bo")
f_ax1.plot(-0.5,np.sqrt(3)/2, "bo")
f_ax1.text(-0.4,np.sqrt(3)/2-0.1,'$P(x,y)$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='center')
f_ax1.text(np.sqrt(2)/2+0.1,np.sqrt(3)/2-0.1,'$t > 0$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='center')
f_ax1.set_xlabel("x")
f_ax1.set_ylabel('y',rotation='horizontal')
f_ax1.axhline(0, color='grey')
f_ax1.axvline(0, color='grey')
f_ax1.grid()
#f_ax1.legend(loc = 'best')

#grafico de la fucnión desplazada hacia arriba
f_ax2 = fig.add_subplot(gs[0, 1])
f_ax2.set_title("$Sentido$ $horario$ $t < 0$ " )
f_ax2.plot(x, y, color="red", markersize=1)
f_ax2.plot(x1, -y1, color="b", markersize=1)
f_ax2.plot(1, 0, "bo")
f_ax2.plot(-0.5,-np.sqrt(3)/2, "bo")
f_ax2.set_xlabel('x')
f_ax2.set_ylabel('y',rotation='horizontal')
f_ax2.text(-0.4,-np.sqrt(3)/2+0.1,'$P(x,y)$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='center')
f_ax2.text(np.sqrt(2)/2 +0.1,-np.sqrt(3)/2+0.1,'$t < 0$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='center')
f_ax2.axhline(0, color='grey')
f_ax2.axvline(0, color='grey')
#f_ax2.legend(loc='upper center')
f_ax2.grid()
```

Si la circunferencia de círculo unitario es $C=2\pi r = 2 \pi 1 = 2 \pi $ entonces si un punto empieza en $(0,1)$ y se desplaza en sentido antihorario a lo largo de toda la circunferencia y regresa al punto $(1,0)$ se desplaza una distancia de $2\pi$. Para desplazarse medio camino alrededor del círculo, recorre una distancia de $\frac{1}{2}* (2\pi)=\pi$. Para moverse un cuarto de de la distancia alrededor del círculo, recorre una distancia de $\frac{1}{2}* (2\pi)=\frac{\pi}{2}$. 

Veamos esto gráficamente.

```python
# nbi:hide_in
%matplotlib inline

plt.rcParams.update({'font.size':fig_font_size})
plt.rcParams.update({'font.family':fig_font_family})


num_segmentos = 200
rad = 1
cx = 0
cy = 0

angulo = np.linspace(0, 2*np.pi, num_segmentos+1)
x = rad * np.cos(angulo) + cx
y = rad * np.sin(angulo) + cy

angulo = np.linspace(0, np.pi/2, num_segmentos+1)
x1 = rad * np.cos(angulo) + cx
y1 = rad * np.sin(angulo) + cy

angulo = np.linspace(0, np.pi, num_segmentos+1)
x2 = rad * np.cos(angulo) + cx
y2 = rad * np.sin(angulo) + cy

angulo = np.linspace(0, (3/2)*np.pi, num_segmentos+1)
x3 = rad * np.cos(angulo) + cx
y3 = rad * np.sin(angulo) + cy


#grafico de las las funciones
fig = plt.figure(figsize=(10,10), dpi= fig_dpi , constrained_layout=True)
gs = fig.add_gridspec(2, 2)

#grafico de la función original
f_ax1 = fig.add_subplot(gs[0, 0])
f_ax1.set_title("$t$ recorre $\pi/2$ (1/4 de vuelta)")
f_ax1.plot(x, y, color="red", markersize=1)
f_ax1.plot(x1, y1, color="b", markersize=1)
f_ax1.plot(1, 0, "bo")
f_ax1.plot(0,1, "bo")
f_ax1.text(0,0.9,'$P(0,1)$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='center', verticalalignment='top')
f_ax1.text(np.sqrt(2)/2,0.9,'$t=\pi / 2}$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='top')
f_ax1.set_xlabel("x")
f_ax1.set_ylabel('y',rotation='horizontal')
f_ax1.axhline(0, color='grey')
f_ax1.axvline(0, color='grey')
f_ax1.grid()
#f_ax1.legend(loc = 'best')

#grafico de la fucnión desplazada hacia arriba
f_ax2 = fig.add_subplot(gs[0, 1])
f_ax2.set_title("$t$ recorre $\pi$ (1/2 de vuelta)" )
f_ax2.plot(x, y, color="red", markersize=1)
f_ax2.plot(x2, y2, color="b", markersize=1)
f_ax2.plot(1, 0, "bo")
f_ax2.plot(-1,0, "bo")
f_ax2.set_xlabel('x')
f_ax2.set_ylabel('y',rotation='horizontal')
f_ax2.text(-0.9,0,'$P(-1,0)$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='center')
f_ax2.text(np.sqrt(2)/2,0.9,'$t= \pi}$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='top')
f_ax2.axhline(0, color='grey')
f_ax2.axvline(0, color='grey')
#f_ax2.legend(loc='upper center')
f_ax2.grid()


#grafico de la función original
f_ax1 = fig.add_subplot(gs[1, 0])
f_ax1.set_title("$t$ recorre $3\pi/2$ (3/2 de vuelta)")
f_ax1.plot(x, y, color="red", markersize=1)
f_ax1.plot(x3, y3, color="b", markersize=1)
f_ax1.plot(1,0, "bo")
f_ax1.plot(0, -1, "bo")
f_ax1.text(0,-0.9,'$P(0, -1)$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='center', verticalalignment='bottom')
f_ax1.text(np.sqrt(2)/2,0.9,'$t=3\pi / 2}$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='top')
f_ax1.set_xlabel("x")
f_ax1.set_ylabel('y',rotation='horizontal')
f_ax1.axhline(0, color='grey')
f_ax1.axvline(0, color='grey')
f_ax1.grid()
#f_ax1.legend(loc = 'best')

#grafico de la fucnión desplazada hacia arriba
f_ax2 = fig.add_subplot(gs[1, 1])
f_ax2.set_title("$t$ recorre $2\pi$ (1 vuelta completa)" )
f_ax2.plot(x, y, color="b", markersize=1)
f_ax2.plot(1, 0, "bo")
f_ax2.set_xlabel('x')
f_ax2.set_ylabel('y',rotation='horizontal')
f_ax2.text(0.9,0,'$P(1,0)$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='right', verticalalignment='center')
f_ax2.text(np.sqrt(2)/2,0.9,'$t = 2\pi}$',fontsize=14,bbox=dict(facecolor='red', alpha=0.3),
                horizontalalignment='left', verticalalignment='top')
f_ax2.axhline(0, color='grey')
f_ax2.axvline(0, color='grey')
#f_ax2.legend(loc='upper center')
f_ax2.grid()
```

Note que ahora utilizando solo el <span style="color:red"><b>parámetro</b></span> <span style="color:red"><b>$t$</b></span> podemos determinar un punto cualquiera de la circunferencia. Esto es la base de las funciones trigonométricas.
